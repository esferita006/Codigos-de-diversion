```python
import turtle


# =============================================================================
# Configuración general
# =============================================================================
START_Y = 200
HEART_FILL = "red"
LEAF_FILL = "dark green"


def setup_turtle(speed: int = 0) -> tuple[turtle.Turtle, turtle.Screen]:
    """
    Crea la pantalla y el 'turtle', define velocidad y estilos básicos.

    Parameters
    ----------
    speed : int
        Velocidad del turtle. 0 = máxima.

    Returns
    -------
    (t, screen) : tuple
        t es el objeto Turtle y screen es la ventana.
    """
    screen = turtle.Screen()
    screen.title("Dibujo Turtle - Corazón y hojas")
    t = turtle.Turtle()
    t.speed(speed)
    t.pensize(2)
    return t, screen


def move_to_start(t: turtle.Turtle, y: float = START_Y) -> None:
    """
    Mueve el turtle a una posición inicial adecuada sin dibujar.
    """
    t.penup()
    t.left(90)
    t.forward(y)
    t.right(90)
    t.pendown()


# =============================================================================
# Dibujo principal (corazón)
# =============================================================================
def draw_heart(t: turtle.Turtle, fill_color: str = HEART_FILL) -> None:
    """
    Dibuja el corazón usando la misma secuencia de movimientos del código original,
    encapsulado y con relleno.

    Nota: se conserva la coreografía exacta para mantener el mismo resultado visual.
    """
    t.fillcolor(fill_color)
    t.begin_fill()

    t.circle(10, 180)
    t.circle(25, 110)
    t.left(50)
    t.circle(60, 45)
    t.circle(20, 170)
    t.right(24)
    t.forward(30)
    t.left(10)
    t.circle(30, 110)
    t.forward(20)
    t.left(40)
    t.circle(90, 70)
    t.circle(30, 150)
    t.right(30)
    t.forward(15)
    t.circle(80, 90)
    t.left(15)
    t.forward(45)
    t.right(165)
    t.forward(20)
    t.left(155)
    t.circle(150, 80)
    t.left(50)
    t.circle(150, 90)

    t.end_fill()


# =============================================================================
# Detalles y hojas (parte verde)
# =============================================================================
def draw_top_details(t: turtle.Turtle) -> None:
    """
    Traza los detalles superiores (sin relleno) que conectan con las hojas.
    """
    t.left(150)
    t.circle(-90, 70)
    t.left(20)
    t.circle(75, 105)
    t.setheading(60)
    t.circle(80, 98)
    t.circle(-90, 40)

    t.left(180)
    t.circle(90, 40)
    t.circle(-80, 98)
    t.setheading(-83)


def draw_leaf_pair_left(t: turtle.Turtle, fill_color: str = LEAF_FILL) -> None:
    """
    Dibuja el primer par de hojas (lado izquierdo) con relleno.
    """
    t.forward(30)
    t.left(90)
    t.forward(25)
    t.left(45)

    t.fillcolor(fill_color)
    t.begin_fill()
    t.circle(-80, 90)
    t.right(90)
    t.circle(-80, 90)
    t.end_fill()

    t.right(135)
    t.forward(60)
    t.left(180)
    t.forward(85)
    t.left(90)
    t.forward(80)


def draw_leaf_pair_right(t: turtle.Turtle, fill_color: str = LEAF_FILL) -> None:
    """
    Dibuja el segundo par de hojas (lado derecho) con relleno.
    """
    t.right(90)
    t.right(45)

    t.fillcolor(fill_color)
    t.begin_fill()
    t.circle(80, 90)
    t.left(90)
    t.circle(80, 90)
    t.end_fill()

    t.left(135)
    t.forward(60)
    t.left(180)
    t.forward(60)
    t.right(90)
    t.circle(200, 60)


# =============================================================================
# Programa principal
# =============================================================================
def main():
    t, screen = setup_turtle(speed=0)

    move_to_start(t, y=START_Y)

    draw_heart(t, fill_color=HEART_FILL)
    draw_top_details(t)
    draw_leaf_pair_left(t, fill_color=LEAF_FILL)
    draw_leaf_pair_right(t, fill_color=LEAF_FILL)

    # Mantener ventana abierta
    screen.mainloop()


if __name__ == "__main__":
    main()
```
